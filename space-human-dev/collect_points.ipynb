```python
LEFT_EYE = [362, 382, 381, 380, 374, 373, 390, 249, 263, 466, 388, 387, 386, 385, 384, 398]
RIGHT_EYE = [33, 7, 163, 144, 145, 153, 154, 155, 133, 173, 157, 158, 159, 160, 161, 246]
LEFT_IRIS = [474, 475, 476, 477]
RIGHT_IRIS = [469, 470, 471, 472]
EYE_CENTER_LEFT = [468]
EYE_CENTER_RIGHT = [473]
```

```python
# Mudança de base
#LEFT_EYE = [x + 1 for x in LEFT_EYE]
#RIGHT_EYE = [x + 1 for x in RIGHT_EYE]
#LEFT_IRIS = [x + 1 for x in LEFT_IRIS]
#RIGHT_IRIS = [x + 1 for x in RIGHT_IRIS]
```

```python
# Defina as listas de índices dos olhos e íris
eyes_and_irises = [LEFT_EYE, RIGHT_EYE, LEFT_IRIS, RIGHT_IRIS, EYE_CENTER_LEFT, EYE_CENTER_RIGHT]

# Crie um dicionário vazio para armazenar os resultados
result = {}

# Loop pelas listas de índices
for eye_or_iris in eyes_and_irises:
    # Loop pelos índices da lista atual
    for index in eye_or_iris:
        # Verifique se o índice está no dicionário
        if index in dicionario:
            # Adicione o valor do dicionário ao resultado
            result[index] = dicionario[index]

# Imprima os resultados
print(result)
```

```python
type(eyes_and_irises)
```

```python
ALL_INDICES = set(LEFT_EYE +  RIGHT_EYE + LEFT_IRIS + RIGHT_IRIS + EYE_CENTER_LEFT + EYE_CENTER_RIGHT)
```

```python
type(ALL_INDICES)
```

```python
def filter_points(all_points, indices):
    return [all_points[i] for i in indices if i < len(all_points)]
```

```python
# Filter points based on the specified indices
points = filter_points(points, ALL_INDICES)
```

```python
len(points)
```

```python
# Create a copy of the frame for drawing
display_frame = image_resize.copy()

# Draw face points
for point in points:
    cv2.circle(display_frame, tuple(point), 2, (0, 255, 0), -1)
```

```python
landmask_image = np.zeros_like(display_frame)
for point in points:
    cv2.circle(landmask_image, tuple(point), 2, (255, 255, 255), -1)
    landmask_image = 255 - landmask_image
```

```python
cv2.imshow('Face Analysis', display_frame)
cv2.imshow('Landmask', landmask_image)         
```

```python
plt.figure()
plt.imshow(landmask_image)
plt.figure()
plt.imshow(display_frame[...,::-1])
plt.show()
```
